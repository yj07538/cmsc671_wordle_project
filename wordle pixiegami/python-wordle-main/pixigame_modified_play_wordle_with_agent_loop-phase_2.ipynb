```python
# def main(): #pg

def main(run_agent = True,run_human = False,play_again = play_again,my_printing = False): #jm

    while(play_again == "Y"):
    
        #word_set = load_word_set("data/wordle_words.txt") #pg
        word_set = load_word_set("data/wordle_la.txt") # this is the reduced possible wordle word list


        np_word_list = np.array(list(word_set)) #jm
        secret = random.choice(list(word_set)) #pg
        wordle = Wordle(secret) #pg
        print(secret) #jm
        my_secret = wordle.secret #jm
        print(my_secret) #jm
        my_wordlist = np_word_list


        # this is the starter word initialization
        starter_alphabet_df = alphabet_frequency(np_word_list) 
        #print(starter_alphabet_df)
        my_starter_list = vocab_value(np_word_list,starter_alphabet_df)
        #print(my_starter_list)
        my_best_starter_word = list(my_starter_list.zipped_lists.items())[0][0]
        #print("my_best_starter_word",my_best_starter_word)


        my_best_guess = my_best_starter_word
        my_attempts = []
        my_file_path = "data/junk_game_file"

        while wordle.can_attempt: #pg

            if run_agent == True: #jm
                agent_guess_list = check_agent_list() #jm
                if len(agent_guess_list) > 0: #jm
                    x = agent_guess_list.pop() #jm

                x = my_best_guess  # this starts with the initial guess from before the can attampt loop
                                   # Then gets the guess from the feedback loop output

                my_results = run_my_agent(my_best_guess,my_secret,my_wordlist,my_file_path,my_printing)
                
                my_wordlist = my_results.myregexresult
                my_len = len(my_wordlist)
                print("my_len",my_len)
                if(my_len > 0):
                    
                    my_best_guess = list(my_results.zipped_lists.items())[0][0]  # this becomes the guess
                    




            if run_human == True:
                x = input("\nType your guess: ") # this is the keyboard request #pg





            x = x.upper()



            if len(x) != wordle.WORD_LENGTH: #pg
                print( #pg
                    Fore.RED #pg
                    + f"Word must be {wordle.WORD_LENGTH} characters long!" #pg
                    + Fore.RESET #pg
                ) #pg
                continue #pg

            if not x in word_set: #pg
                print( #pg
                    Fore.RED #pg
                    + f"{x} is not a valid word! this word is not in the word set" #pg
                    + Fore.RESET #pg
                ) #pg
                continue #pg

            wordle.attempt(x) #pg
            my_attempts = wordle.attempts
            print("display_results- wordle.attempts:",my_attempts) #jm

            display_results(wordle) #pg






        if wordle.is_solved: #pg
            print("You've solved the puzzle.") #pg
        else: #pg
            print("You failed to solve the puzzle!") #pg
            print(f"The secret word was: {wordle.secret}") #pg


        play_again = input("Want to play again?  Y/N") 
    
print("Thank you for playing!")
    

```

```python
def display_results(wordle: Wordle): #pg
    print("\nYour results so far...") #pg
    print(f"You have {wordle.remaining_attempts} attempts remaining.\n") #pg

    lines = [] #pg

    print("display_results- wordle.attempts:",wordle.attempts) #jm

    
    for word in wordle.attempts: #pg
        print("display_results- word:",word) #pg
        result = wordle.guess(word) #pg
        colored_result_str = convert_result_to_color(result) #pg
        lines.append(colored_result_str) #pg

    for _ in range(wordle.remaining_attempts): #pg
        lines.append(" ".join(["_"] * wordle.WORD_LENGTH)) #pg

    draw_border_around(lines) #pg
```

```python
def load_word_set(path: str): #pg
    word_set = set() #pg
    with open(path, "r") as f: #pg
        for line in f.readlines(): #pg
            word = line.strip().upper() #pg
            word_set.add(word) #pg
    return word_set #pg
```

```python
# this loads guesses from the agent guess file into a list
def check_agent_list(): #jm
    my_set = load_word_set("data/agent_guesses.txt")
    mylist = list(my_set)
    print(mylist, type(mylist))
    return(mylist)
```

```python
def convert_result_to_color(result: List[LetterState]): #pg #jm this has been modified with background letters
    result_with_color = []  #pg
    for letter in result: #pg
        if letter.is_in_position: #pg
            #color = Fore.GREEN #pg
            color = Back.GREEN #jm
        elif letter.is_in_word: #pg
            #color = Fore.YELLOW #pg
            color = Back.YELLOW #jm
        else: #pg
            color = Fore.WHITE #pg
        colored_letter = color + Fore.BLACK + letter.character + Fore.RESET + Back.RESET #jm
        result_with_color.append(colored_letter) #pg
    return " ".join(result_with_color) #pg
```

```python
def draw_border_around(lines: List[str], size: int = 9, pad: int = 1): #pg

    content_length = size + pad * 2
    top_border = "┌" + "─" * content_length + "┐"
    bottom_border = "└" + "─" * content_length + "┘"
    space = " " * pad
    print(top_border)

    for line in lines:
        print("│" + space + line + space + "│")

    print(bottom_border)


```

```python
def write_agent_guess_set(path: str, my_list: List[str]): #pg
    word_set = set()
    with open(path, "w") as f:
        f.writelines("%s\n" % l for l in my_list)
```
